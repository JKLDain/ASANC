## Features

```python
# Tempo(BPM)
def Bpm(y): 
    bpm = []
    for i in range(len(y)):
        tempo, _ = librosa.beat.beat_track(y[i])
        bpm.append(tempo)
    return bpm

# Zero Crossing Rate (음파가 양에서 음으로 또는 음에서 양으로 바뀌는 비율)
def Zcr(y):
    zcr=[]
    for i in range(len(y)):
        zero_crossing_rate = librosa.feature.zero_crossing_rate(y[i],pad=False)
        zcr.append(zero_crossing_rate)
    return zcr

#Harmonic : 사람의 귀로 구분할 수 없는 특징들(오디오의 색깔)
def HPC(y):
    harm = []
    perc = []
    for i in range(len(y)):
        y_harm, y_perc = librosa.effects.hpss(y[i])
        harm.append(y_harm)
    return harm

# Spectral Centroid : 소리를 주파수로 표현했을 때, 주파수의 가중평균을 계산하여 소리의 "무게중심"이 어딘지를 알려주는 지표
def SC(y):
    sc = []
    for i in range(len(y)):
        spectral_centroids = librosa.feature.spectral_centroid(y[i])
        sc.append(spectral_centroids)
    return sc

# 신호 모양 측정 , 총 스펙트럴 에너지 중 낮은 주파수(85% 이하)에 얼마나 많이 집중되어 있는가
def SpectralRolloff(y): 
    rolloff = []
    for i in range(len(y)):
        spectral_rolloff = librosa.feature.spectral_rolloff(y[i])
        rolloff.append(spectral_rolloff)
    return rolloff    

# MFCCs는 특징들의 작은 집합(약10-20)으로 스펙트럴 포곡선의 전체적인 모양을 축약하여 보여준다
# 사람의 청각구조를 반영하여 음성 정보 추출 
def MFCCs(y): 
    mfccs = []
    for i in range(len(y)):
        mfcc = librosa.feature.mfcc(y[i])
        mfcc = sklearn.preprocessing.minmax_scale(mfcc, axis=1)
        mfccs.append(mfcc)
    return mfccs

#크로마 특징은 음악의 흥미롭고 강렬한 표현이다.
def ChromaFrequencies(y):
    cf = []
    for i in range(len(y)):
        chromagram = librosa.feature.chroma_stft(y[i],hop_length = 512)
        cf.append(chromagram)
    return cf

def mean(x):
    means = []
    for i in range(len(x)):
        means.append(np.mean(x[i]))
    return means

def var(x):
    variances = []
    for i in range(len(x)):
        variances.append(np.var(x[i]))
    return variances



```
